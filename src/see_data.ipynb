```python
import torch
from safetensors.torch import load_file
import os

ds_base = "/home/harry/ssd1/data/eulerCart/box_1024-1"

files = list(filter(lambda s: s.endswith(".safetensors"), os.listdir(ds_base)))

files = sorted(files)

files = [os.path.join(ds_base, file) for file in files]
```

```python
fileC = files[-5000]

dataC = load_file(fileC)

display(dataC["u"].shape)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig.set_size_inches((12, 11))

ax.cla()
ax.pcolormesh(dataC["u"][:,:,0].cpu(), shading="auto", cmap="jet")
ax.axis("equal")
plt.show(block=False)
```
